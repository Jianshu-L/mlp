```python
data = np.load('Adam_4layers_100_L2.npy')
data1 = np.load('Adam_4layers_200_L2_200.npy')
data2 = np.load('Adam_4layers_300_L2_200.npy')

fig_1 = plt.figure(figsize=(10,20))

ax_1 = fig_1.add_subplot(411)
ax_1.plot(np.arange(data[0].shape[0]), data[0], label = '100')
ax_1.plot(np.arange(data1[0].shape[0]), data1[0], label = '200')
ax_1.plot(np.arange(data2[0].shape[0]), data2[0], label = '300')
ax_1.legend(loc=0)
ax_1.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_1.set_ylabel('Training set error', fontsize = 12, fontweight = 1000)


ax_2 = fig_1.add_subplot(412)
ax_2.plot(np.arange(data[2].shape[0]), data[2], label = '100')
ax_2.plot(np.arange(data1[2].shape[0]), data1[2], label = '200')
ax_2.plot(np.arange(data2[2].shape[0]), data2[2], label = '300')
ax_2.legend(loc=0)
ax_2.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_2.set_ylabel('Validation set error', fontsize = 12, fontweight = 1000)

ax_3 = fig_1.add_subplot(413)
ax_3.plot(np.arange(data[1].shape[0]), data[1], label = '100')
ax_3.plot(np.arange(data1[1].shape[0]), data1[1], label = '200')
ax_3.plot(np.arange(data2[1].shape[0]), data2[1], label = '300')
ax_3.legend(loc=0)
ax_3.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_3.set_ylabel('Training set ACC', fontsize = 12, fontweight = 1000)


ax_4 = fig_1.add_subplot(414)
ax_4.plot(np.arange(data[3].shape[0]), data[3], label = '100')
ax_4.plot(np.arange(data1[3].shape[0]), data1[3], label = '200')
ax_4.plot(np.arange(data2[3].shape[0]), data2[3], label = '300')
ax_4.legend(loc=0)
ax_4.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_4.set_ylabel('Validation set ACC', fontsize = 12, fontweight = 1000)

plt.tight_layout()
fig_1.savefig('Adam_4layers_10.pdf')
print(data[2][-1], data[3][-1])
print(data1[2][-1], data1[3][-1])
print(data2[2][-1], data2[3][-1])
```
