```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
inputs = tf.placeholder(tf.float32, [None,120,25,1], 'inputs')
targets = tf.placeholder(tf.float32, [None, 10], 'targets')
```

```python
data = np.load('LSTM_1layers_200.npy')
data1 = np.load('GRU_1layers_200.npy')

fig_1 = plt.figure(figsize=(10,20))

ax_1 = fig_1.add_subplot(411)
ax_1.plot(np.arange(data[0].shape[0]), data[0], label = 'LSTM')
ax_1.plot(np.arange(data1[0].shape[0]), data1[0], label = 'GRU')
ax_1.legend(loc=0)
ax_1.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_1.set_ylabel('Training set error', fontsize = 12, fontweight = 1000)


ax_2 = fig_1.add_subplot(412)
ax_2.plot(np.arange(data[2].shape[0]), data[2], label = 'LSTM')
ax_2.plot(np.arange(data1[2].shape[0]), data1[2], label = 'GRU')
ax_2.legend(loc=0)
ax_2.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_2.set_ylabel('Validation set error', fontsize = 12, fontweight = 1000)

ax_3 = fig_1.add_subplot(413)
ax_3.plot(np.arange(data[1].shape[0]), data[1], label = 'LSTM')
ax_3.plot(np.arange(data1[1].shape[0]), data1[1], label = 'GRU')
ax_3.legend(loc=0)
ax_3.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_3.set_ylabel('Training set ACC', fontsize = 12, fontweight = 1000)


ax_4 = fig_1.add_subplot(414)
ax_4.plot(np.arange(data[3].shape[0]), data[3], label = 'LSTM')
ax_4.plot(np.arange(data1[3].shape[0]), data1[3], label = 'GRU')
ax_4.legend(loc=0)
ax_4.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_4.set_ylabel('Validation set ACC', fontsize = 12, fontweight = 1000)

plt.tight_layout()
fig_1.savefig('RNN_layers_10.pdf')

print(data[2][-1], data[3][-1])
print(data1[2][-1], data1[3][-1])
```
