```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
inputs = tf.placeholder(tf.float32, [None,120,25,1], 'inputs')
targets = tf.placeholder(tf.float32, [None, 10], 'targets')
```

```python

data1 = np.load('GRU_1layers_200_l2_10_3.npy')
data2 = np.load('GRU_1layers_200_l2_10_4.npy')

fig_1 = plt.figure(figsize=(10,20))

ax_1 = fig_1.add_subplot(411)

ax_1.plot(np.arange(data1[0].shape[0]), data1[0], label = '0.001')
ax_1.plot(np.arange(data2[0].shape[0]), data1[0], label = '0.0001')
ax_1.legend(loc=0)
ax_1.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_1.set_ylabel('Training set error', fontsize = 12, fontweight = 1000)


ax_2 = fig_1.add_subplot(412)

ax_2.plot(np.arange(data1[2].shape[0]), data1[2], label = '0.001')
ax_2.plot(np.arange(data2[2].shape[0]), data1[2], label = '0.0001')
ax_2.legend(loc=0)
ax_2.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_2.set_ylabel('Validation set error', fontsize = 12, fontweight = 1000)

ax_3 = fig_1.add_subplot(413)

ax_3.plot(np.arange(data1[1].shape[0]), data1[1], label = '0.001')
ax_3.plot(np.arange(data2[1].shape[0]), data1[1], label = '0.0001')
ax_3.legend(loc=0)
ax_3.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_3.set_ylabel('Training set ACC', fontsize = 12, fontweight = 1000)


ax_4 = fig_1.add_subplot(414)

ax_4.plot(np.arange(data1[3].shape[0]), data1[3], label = '0.001')
ax_4.plot(np.arange(data2[3].shape[0]), data1[3], label = '0.0001')
ax_4.legend(loc=0)
ax_4.set_xlabel('Epoch number',fontsize = 12, fontweight = 1000)
ax_4.set_ylabel('Validation set ACC', fontsize = 12, fontweight = 1000)

plt.tight_layout()



print(data1[2][-1], data1[3][-1])
print(data2[2][-1], data2[3][-1])
```
